```python
import requests
import os

models = ['ACCESS-CM2']
assembly_member = ['r2i1p1f1'] 
var = ['tasmax']
ssps = ['ssp245']


for model,member in zip(models,assembly_member):
    for ssp in ssps:
        for variable in var:
            #if (model == 'ACCESS-ESM1-5') and (variable=='pr' or variable=='hurs' or variable=='tasmax'):
            #    continue
            output_dir = f"D:\\Jake_ClimateRasters\\QLD_Future_Climate\\{model}_{ssp}_{member}\\{variable}\\"
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)
            year = 2036
            while year <= 2036:
                # Download for a specific area north=-9&west=138&east=155&south=-29.5, Chnage coordinates in the address or leave out for all of Australia
                ncss_url = f"https://thredds.nci.org.au/thredds/ncss/grid/ig45/QldFCP-2/output/CMIP6/DD/AUS-10i/UQ-DEC/{model}/{ssp}/{member}/CCAMoc-v2112/v1-r1/day/{variable}/latest/{variable}_AUS-10i_{model}_{ssp}_{member}_UQ-DEC_CCAMoc-v2112_v1-r1_day_{year}0101-{year}1231.nc?var={variable}&horizStride=1&time_start={year}-01-01T12:00:00Z&time_end={year}-12-31T12:00:00Z&&&accept=netcdf4-classic"
                response = requests.get(ncss_url)
                if response.status_code == 200:
                    output_file = os.path.join(output_dir, f"{variable}_AUS-10i_{model}_{ssp}_{member}_UQ-DEC_CCAMoc-v2112_v1-r1_day_{year}0101-{year}1231.nc")
                    with open(output_file, "wb") as f:
                        f.write(response.content)
                    print(f"{model}, {member}, SSP{ssp}, {variable}, {year}: Download successful")
                else:
                    print(f"Failed to download data: {response.status_code}")
                year += 1
```

```python
# Process clim data to get correct time averaging in the catchment
# Normalise the time axis to be consistent with the other data from "days since" to actual date.
import xarray as xr
import pandas as pd
import os
import cftime

models = ['ACCESS-ESM1-5','GISS-E2-1-G','EC-Earth3']#
assembly_member = ['r6i1p1f1','r2i1p1f2','r1i1p1f1'] 
institutions = ['UQ-DEC_CCAMoc-v2112','ALLUVIUM-JAN_FAO56-PenmenMonteith']
var = ['pr','et0']

for model,member in zip(models,assembly_member):
    for variable, inst in zip(var,institutions):
        output_dir = f'C:\\QLD_Future_Climate\\{model}_ssp370_{member}\\{variable}_norm\\'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        year = 2015
        while year <= 2100:
            netcdf_file = f'C:\\QLD_Future_Climate\\{model}_ssp370_{member}\\{variable}\\{variable}_AUS-10i_{model}_ssp370_{member}_{inst}_v1-r1_day_{year}0101-{year}1231.nc'
            ds = xr.open_dataset(netcdf_file)
            time_index = ds['time'].to_index()
            datetime_index = pd.DatetimeIndex([pd.Timestamp(t.strftime('%Y-%m-%d')) for t in time_index])
            # Assign the new DatetimeIndex back to the dataset
            ds['time'] = datetime_index
            ds.to_netcdf(output_dir + f"{variable}_AUS-10i_{model}_ssp370_{member}_{inst}_v1-r1_day_{year}0101-{year}1231.nc")
            print(f"{model}, {member}, {variable}, {year}: Normalization successful")
            year += 1
```
